### Introduction

There are two main ways to create plots. The first way uses a global state machine to interact directly with Matplotlib's pyplot module. The second, object-oriented approach uses figure and axes classes to provide a programmatic alternative.

Generally, if working interactively with single plots, pyplot's global state is a convenient shortcut. For all other occasions, it makes sense to explicitly declare figures and axes using the object-oriented approach.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Creates pandas datetime index with 200 day (d) elements, starting form the current time (datetime.now())
x = pd.period_range(datetime.now(), periods=200, freq='d')

# Converts datetime index to Python datetimes
x = x.to_timestamp().to_pydatetime()

# Creates three 200-element random arrays summed along the 0 axis
y = np.random.randn(200, 3).cumsum(0)
```

We now have a y-axis with 200 time slots and three random array sfor the complementary x values. These are provided as separate arguments to the (line)plot method.

```python
plt.plot(x, y)
```

### Configurations

Matplotlib provides a wide range of configurations, which can be specified in a matplotlibrc file or dynamically through the dictionary-like rcParams variable.

```python
import matplotlib as mpl

# Changes the width and default color of plot lines
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r' # red
```

As well as using the rcParams variable, you can use the gcf (get current figure) method to grab the currently active figure and manipulate it directly.

```python
# Two ways to set the figure size to 8 by 4 inches
plt.rcParams['figure.figsize'] = (8, 4)
plt.gcf().set_size_inches(8, 4)
```

### Points, Not Pixels

Matplotlib uses points, not pixels, to measure the size of its figures. By default, a point is approximately 1/72 of an inch wide, but Matplotlib allows you to adjust this by changing the dots-per-inch (dpi) for any figures generated. The higher this number, the better quality of the image.

### Full Customized Line Chart

```python
plots = plt.plot(x, y)
plt.legend(plots, ('foo', 'bar', 'baz'), loc='best',
                    framealpha=0.25,
                    prop={'size':'small', 'family':'monospace'})
plt.gcf().set_size_inches(8, 4)
plt.title('Random trends')
plt.xlabel('Date')
plt.ylabel('Cum. sum')
plt.grid(True) 
plt.figtext(0.995, 0.01, '© Acme Designs 2021',
ha='right', va='bottom')
# Should guarantee that all plot elements are within figure box
plt.tight_layout()
```
